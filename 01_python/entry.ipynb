你可以假设每种输入只会对应一个答案。但是，数组中同一个元素在答案里不能重复出现。


来源： leetcode - https://leetcode-cn.com/problems/two-sum/



````python
inputs = [
    {
        "nums": [1, 2, 3],
        "target": 3,
        "answer": [0, 1]
    },
    {
        "nums": [2, 7, 11, 15],
        "target": 9,
        "answer": [0, 1]
    },
    {
        "nums": [5, 5, 10, 11],
        "target": 15,
        "answer": [0, 2]
    },
     {
        "nums": [1, 2, 3 ,4, 5],
        "target": 8,
        "answer": [2, 4]
    },
    {
        "nums": [10, 11, 22, 44],
        "target": 55,
        "answer": [1, 3]
    },
]

def yourcode(nums, target):
    # 请完善代码
    # i = 0
    # j = i + 1
    # while i < len(nums):
    #     nums_i = nums[i]
    #     while j < len(nums):
    #         nums_j = nums[j]
    #         if nums_i + nums_j == target:
    #             return [i, j]
    #         j += 1
    #     i += 1
    #     j = i + 1
    for i in range(len(nums)):
        nums_i = nums[i]
        for j in range(i, len(nums)):
            nums_j = nums[j]
            if nums_i + nums_j == target:
                return [i, j]
    
for x in inputs:
    nums, target, answer = x['nums'], x['target'], x['answer']
    your_answer = yourcode(nums, target)
    print(answer == your_answer, answer, your_answer)
    
    
````

## 随堂练习2
给你一个整数 `x` ，如果 `x` 是一个回文整数，返回 `True` ；否则，返回 `False` 。

回文数是指正序（从左向右）和倒序（从右向左）读都是一样的整数。

例如，121 是回文，而 123 不是。

来源：力扣（LeetCode） - https://leetcode-cn.com/problems/palindrome-number

````python
inputs = [
    {
        "x": 131,
        "answer": True
    },
    {
        "x": 432,
        "answer": False
    },
    {
        "x": 12345677654321,
        "answer": True
    },
    {
        "x": 445544,
        "answer": True
    },
    
]

def yourcode(x):
    # 请完善代码
    x_str = str(x)
    x_rev_str = list(reversed(x_str))
    for i in range(len(x_str)):
        print(x_rev_str[i], x_str[i])
        if x_rev_str[i] != x_str[i]:
            return False
    return True
    
for i in inputs:
    x, answer = i['x'], i['answer']
    your_answer = yourcode(x)
    print(answer == your_answer,answer,your_answer)
    
    
````
